```python
import matplotlib.pyplot as plt

# 创建数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 创建简单折线图
plt.plot(x, y, label='Line 1',linewidth=5.0, color='blue', linestyle='solid', marker='<')
# label: 图例标签
# color: 线条颜色
# linestyle: 线型 (可选：'solid', 'dashed', 'dashdot', 'dotted') 或者 ('-', '--', '-.', ':')
# marker: 数据点的形状 (可选：'o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X')

# 添加标题和标签
plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# 显示图例
plt.legend()

# 显示图表
plt.show()
```

```python
# 散点图
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.rand(50) * 1000

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()  # 显示颜色条
plt.title("Scatter Plot Example")
plt.show()
```

```python
# 直方图
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 8, 5]

plt.bar(categories, values, color='purple')
plt.title("Bar Plot Example")
plt.show()
```

```python
# 饼图
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
```
