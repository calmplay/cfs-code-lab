```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
values = [3, 7, 8, 5]
data = np.random.randn(1000)
categories = ['A', 'B', 'C', 'D']
labels = ['A', 'B', 'C', 'D']
```

```python
# 子图
# fig, axs = plt.subplots(2, 2, figsize=(10, 10))
_, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(x, y, 'r')  # 红线折线图
axs[0, 0].set_title("Subplot 1")

axs[0, 1].scatter(x, y)  # 散点图
axs[0, 1].set_title("Subplot 2")

axs[1, 0].bar(categories, values)  # 柱状图
axs[1, 0].set_title("Subplot 3")

axs[1, 1].hist(data, bins=20, color='g')  # 直方图
axs[1, 1].set_title("Subplot 4")

plt.tight_layout()

# 可以通过以下方式更改图形样式
plt.style.use('ggplot')  # 样式包括 'ggplot', 'classic', 'dark_background' 等

# 保存图形 (保存完也会show(),但是show完再保存就是空白)
plt.savefig("plot.png", dpi=300, bbox_inches='tight')
# plt.show()
```

```python
# 添加注释
plt.plot(x, y)
plt.annotate('Max Value', xy=(4, 16), xytext=(6, 20),
             arrowprops=dict(facecolor='yellow', shrink=0.05))
# plt.show()
plt.savefig("plot_ano.png", dpi=300, bbox_inches='tight')
```

```python
### 综合示例
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='red')

plt.title("Sin and Cos Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5)  # 添加x轴
plt.axvline(0, color='black', linewidth=0.5)  # 添加y轴

plt.legend()
plt.grid(True)
plt.show()
```
