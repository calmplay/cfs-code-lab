```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
### 基础 3D 图形
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # 创建一个3D子图

# 数据
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
z = [5, 15, 25, 35, 45]

# 绘制3D散点图
ax.scatter(x, y, z, c='r', marker='o')

# 设置标题和标签
ax.set_title("3D Scatter Plot")
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_zlabel("Z Label")

plt.show()
```

```python
### 高阶3D功能
# 3D 曲线
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 参数化曲线
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t

ax.plot(x, y, z, label='3D Curve', color='blue')
ax.legend()
plt.show()
```

```python
# 3D 曲面
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 创建网格
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # 曲面函数

# 绘制3D曲面
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surf)  # 添加颜色条
plt.show()
```

```python
# 3D 条形图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 数据
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
z = np.zeros(5)

dx = np.ones(5)
dy = np.ones(5)
dz = [1, 2, 3, 4, 5]

# 绘制条形图
ax.bar3d(x, y, z, dx, dy, dz, color='cyan')
plt.show()
```

```python
# 等高线图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 网格
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# 绘制等高线图
ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_title("3D Contour Plot")
plt.show()
```

```python
# 多图层叠加
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 曲面
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z1 = np.sin(np.sqrt(X**2 + Y**2))
Z2 = np.cos(np.sqrt(X**2 + Y**2))

# 绘制两层曲面
ax.plot_surface(X, Y, Z1, cmap='viridis', alpha=0.7)
ax.plot_surface(X, Y, Z2, cmap='plasma', alpha=0.7)

plt.show()
```

```python
# 动态视角
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 曲面
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')  
# alpha设置透明度,值越小越透明; edgecolor设置网格线颜色

# 动态旋转视角
ax.view_init(elev=30, azim=45)  # 调整高度角和方位角
plt.show()
```

```python
# 复杂曲面细节控制
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 曲面
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2

# 细节控制
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)  # 缩小颜色条
plt.show()
# plt.savefig("3d_plot.png", dpi=300)  # 高分辨率输出
```
