```python
# 用()和[]都行
n1 = np.array([1, 2, 3])
n2 = np.array(((1, 2, 3), (4, 5, 6)))
n3 = np.array([[1, 2, 3], [4, 5, 6]])
n1, n2, n3
```

广播机制

```python
c = np.array(93)  # 单个数
a,c
```

```python
a + c, a - c, a * c, a / c
```

```python
a_origin = np.copy(a)
b = np.copy(a)
a[np.arange(2), 1] += 100  # <=> a[[0,1],1]触发广播 => a[0,1] += 100, a[1,1] += 100
b[np.arange(2), np.arange(2)] += 100  # 一一对应 => a[0,0] += 100, a[1,1] += 100
a_origin, a, b
```

```python
a > 10
```

转换类型

```python
# `np.string_` was removed in the NumPy 2.0 release. Use `np.bytes_` instead.
a = np.array([[-1, 0, 3], [4, 5, 6]], dtype=np.bytes_)
(a, a.dtype, a.astype(np.int32), a.astype(np.float32),
 a.astype(np.float32).astype(np.bytes_))
#, a.astype(np.bool) # 可以转会警告,=0为False
```
