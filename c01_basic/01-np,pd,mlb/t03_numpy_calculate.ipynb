数组运算

```python
import numpy as np

seed = 2025
rng = np.random.default_rng(seed)
x = rng.random((2, 5))
x, np.sum(x), np.average(x), np.mean(x)
```

整块复制

```python
np.tile(x, (2, 2))
```

.argsort(),将元素按照行从小到大排序，返回对应位置元素的下标

```python
x.argsort()
```

```python
a = np.array([[[1, 6, 1], [3, 4, 5]], [[99, 6, 1], [3, 4, 5]]])
np.argsort(a, axis=0), np.argsort(a, axis=1), np.argsort(a, axis=2)
# 形状不变
```

row_stack按行拼接 (扩充行)，column_stack (扩充列)  (这是旧版, 不推荐)

```python
a = np.array([[1, 2, 3], [4, 5, 6.0]])
b = np.array([[7, 8, 9], [10, 11, 12]])
np.row_stack((a, b)), np.column_stack((a, b))
```

新版使用 vstack和hstack

```python
np.vstack((a, b)), np.hstack((a, b))
```

逐元素计算

```python
a + b, a - b, a * b, a / b
```

矩阵乘法 (记得保证col_a = row_b) 三种写法

```python
# 下面两种写法等价, 返回新的ndarray
np.dot(a, b.T), a @ b.T
```

```python
# 第三种写法, 先转换为矩阵类型, 这里*就是矩阵乘法了,不过注意: 返回的是矩阵matrix类型
# np.mat(a) * np.mat(b.T) # AttributeError: `np.mat` was removed in the NumPy 2.0 release. Use `np.asmatrix` instead.
np.asmatrix(a) * np.asmatrix(b.T)
```

求逆 (输入若不可逆会报错Singular matrix)

```python
np.linalg.inv(a @ a.T), np.linalg.inv(np.array([[1]]))
# np.linalg.inv(np.array([[1, 0], [0, 0]])))
```

np.argmax 返回最大值的下标

```python
a = np.array([[[1, 6, 1], [3, 4, 5]], [[99, 6, 1], [3, 4, 5]]])
np.argmax(a, axis=0), np.argmax(a, axis=1), np.argmax(a, axis=2)
# 总共0,1,2三维, axis=多少就是把哪个维度消掉
# 比如axis=0,则shape(2,2,3)->(2,3)
# 比如axis=1,则shape(2,2,3)->(2,3)
# 比如axis=2,则shape(2,2,3)->(2,2)
```
