```python
import torch
torch.__version__
```

```python
device = ""
if torch.cuda.is_available():
    device = "cuda"
elif torch.backends.mps.is_available():
    device = "mps"  # macOS m1 的 mps ≈ NVIDIA 1050Ti
else:
    device = "cpu"
# 每次setting里更改本地或远程的解释器, 看这里是cuda还是mps就知道有没有切成功了
# 现在每次都得重新打开项目,才切换
print(f"Using {device} device")
```

```python
t0 = torch.tensor(1)
t1 = torch.tensor([[1, 2, 3], [2, 0, 0]])
t0.dtype,t1.dtype  # 默认都是long(即int64)
```

```python
# 可以手动赋值创建
x = torch.tensor([[1, 2, 3], [2, 0, 0.]])
y = torch.tensor([[1, 2, 3], [2, 0, 0.9]]).type(torch.long)
x, x.shape, x.dtype,y, y.shape, y.dtype
```
