tensor连结

```python
X = torch.arange(12, dtype=torch.float32).reshape((3, 4))
Y = torch.tensor(([2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]))
X, Y, torch.cat((X, Y), dim=0), torch.cat((X, Y), dim=1)
# dim=0 表示在行上拼接(增加行),dim=1 表示在列上拼接(增加列)
```

通过逻辑运算符构建二元张量

```python
X == Y, X > Y, X < Y
```

聚合函数, 生成单元素tensor

```python
X.sum(), X.int(), Y.prod()
```

广播机制(broadcast mechanism)按元素操作

```python
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

```python
a + b  # 维度一致,均扩成3x2,缺失的行列复制已有的
```

```python
X, X[1, 2], X[-1], X[1:3]
```

元素赋值

```python
X[0:2, :] = 666
X[1, 2] = 777
X
```

```python
# 一些操作会导致为新结果分配内存
before = id(Y)
Y = Y + X  # 如果用Y[:]=Y+X或者Y+=X , 则不会重新分配内存
id(Y) == before
```

原地操作

```python
Z1 = torch.zeros_like(Y)
Z2 = torch.zeros_like(Y)
before1 = id(Z1)
before2 = id(Z2)
Z1[:] = Z1 + X
Z2 += X
id(Z1) == before1, id(Z2) == before2
```
