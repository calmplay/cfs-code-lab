```python
import torch

# 创建张量, 这三个变量作为分母
x = torch.randn(size=(), requires_grad=True)  # 标量()
v = torch.randn(2, requires_grad=True) # 向量(2) (默认看作是一个行向量)
v1 = torch.randn(2,1, requires_grad=True)  # 矩阵模拟行向量 (2,1)
v2 = torch.randn(1,2, requires_grad=True)  # 矩阵模拟列向量 (1,2)
M = torch.randn(3,4, requires_grad=True)  # 矩阵 (3, 2)
x,x.shape, v1.shape,v2.shape,M.shape
```

```python
y_x = 2*x
y_x.backward()
print(f"标量对标量求导,shape:{x.grad.shape}, 值{x.grad}", )

x.grad.zero_()
y_x = (x*v1).sum()
y_x.backward()
print(f"行向量对标量求导,shape:{x.grad.shape}, 值{x.grad}", )
```

```python
# 定义一个函数
def f(x):
    return torch.stack([x[0]**2 + x[1], x[0] * x[1]])

# 创建输入张量
x = torch.tensor([1.0, 2.0,3.0], requires_grad=True)

# 计算雅可比矩阵
jacobian_matrix = torch.autograd.functional.jacobian(f, x)

print("雅可比矩阵：")
print(jacobian_matrix)
```
