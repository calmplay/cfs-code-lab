```python
import torch
import torch.nn as nn

softmax = nn.Softmax()
sigmoid = nn.Sigmoid()
relu = nn.ReLU()
leaky_relu = nn.LeakyReLU()

(
    softmax(torch.tensor(3.0)),
    softmax(torch.tensor([[0, 0, 1], [1, 0, 2.0]])),
    softmax(torch.tensor([[2, 1, 1], [1, 3, 3.0]])),
    softmax(torch.tensor([[20, 10, 10], [10, 30, 30.0]])),
    torch.tensor([[3.0, 9999], [123, 234]]).shape,
    softmax(torch.tensor([[3.0, 9999], [123, 234]])),
    softmax(torch.tensor([[3.0, 222], [123, 234]])),
    softmax(torch.tensor([[3.0, 3], [233, 234]])),
    softmax(torch.tensor([[3.0, 9999], [233, 234]])),
    softmax(torch.tensor([[[3.0, 9999], [233, 234]]])),
    torch.tensor([[[233, 234], [233, 234]], [[3.0, 9999], [233, 234]]]).shape,
    softmax(torch.tensor([[[233, 234], [233, 234]], [[3.0, 9999], [233, 234]]])),
)
```

$$
\operatorname{softmax}(\operatorname{input}[0])=\frac{e^{\operatorname{logit}}}{\sum e^{\log \mid t s}}=\left[\frac{e^0}{e^0+e^0+e^1}, \frac{e^0}{e^0+e^0+e^1}, \frac{e^1}{e^0+e^0+e^1}\right]
$$

```python
(
    sigmoid(torch.Tensor([0])),
    sigmoid(torch.Tensor([2])),
    sigmoid(torch.Tensor([999])),
    sigmoid(torch.Tensor([0, 2, 999]))
)
```

```python
(relu(torch.Tensor([-1, 2, 999])).int(),
 relu(torch.Tensor([-99, 2, 999])).int(),
 leaky_relu(torch.Tensor([-1, 2, 999])).int(),
 leaky_relu(torch.Tensor([-999, 2, 999])).int(),
 leaky_relu(torch.Tensor([-99999, 2, 999])).int(),
 )
```
