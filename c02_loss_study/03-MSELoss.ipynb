# MSELoss 均方误差

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

$$
\mathrm{MSE}=\frac{1}{N} \sum_{i=1}^N\left(y_i-\hat{y}_i\right)^2
$$

```python
pred = torch.tensor([2.5, 0.0, 2.0, 8.0])
target = torch.tensor([3.0, -0.5, 2.0, 7.0])

loss_fn = nn.MSELoss()
loss = loss_fn(pred, target)

print("MSE Loss:", loss.item())
```

$\frac{(2.5-3)^2+(0+0.5)^2+(2-2)^2+(8-7)^2}{4}=0.375$

```python
# 调试每个元素的误差
errors = (pred - target) ** 2
print("Squared Errors:", errors)
print("Mean Squared Error:", errors.mean())
```

```python
# 真值固定 = 1.0
# 模型输出从 -3 到 +5，看看 loss 怎么变化
y_true = 1.0
y_pred_range = torch.linspace(-3, 5, 200)
mse_loss = (y_pred_range - y_true) ** 2  # MSE: (y_pred - y_true)^2

plt.plot(y_pred_range.numpy(), mse_loss.numpy())
plt.title("MSE Loss vs Prediction")
plt.xlabel("Prediction")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.show()
```
