# MAE(L1Loss) 绝对误差，Mean Absolute Error

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

$$
\text { L1Loss }=\frac{1}{N} \sum_{i=1}^N\left|y_i-\hat{y}_i\right|
$$

```python
pred = torch.tensor([2.5, 0.0, 2.0, 8.0])
target = torch.tensor([3.0, -0.5, 2.0, 7.0])

loss_fn = nn.L1Loss()
loss = loss_fn(pred, target)

# |2.5-3.0|
print("L1 Loss:", loss.item())
```

$\frac{|2.5-3.0|+|0.0+0.5|+|2-2|+|8-7|}{4}=0.5$

```python
y_true = 1.0
y_pred_range = torch.linspace(-3, 5, 200)
l1_loss = torch.abs(y_pred_range - y_true)

plt.plot(y_pred_range.numpy(), l1_loss.numpy(), label="L1 Loss")
plt.title("L1 Loss vs Prediction")
plt.xlabel("Prediction")
plt.ylabel("L1 Loss")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 对比 MSE 和 L1（差异视觉化）
mse_loss = (y_pred_range - y_true) ** 2

plt.plot(y_pred_range.numpy(), mse_loss.numpy(), label="MSE Loss")
plt.plot(y_pred_range.numpy(), l1_loss.numpy(), label="L1 Loss")
plt.title("MSE vs L1 Loss")
plt.xlabel("Prediction")
plt.ylabel("Loss")
plt.grid(True)
plt.legend()
plt.show()
```

可以看到 MSE 对大偏差增长更快，L1 更“线性稳定”
